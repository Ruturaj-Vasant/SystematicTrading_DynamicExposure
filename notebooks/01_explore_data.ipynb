```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
data = pd.read_csv('../data/spy_vix_data.csv', index_col=0, parse_dates=True)
```

```python
print("✅ Data Info")
print(data.info())
print("\n✅ Data Head")
print(data.head())
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['SPY_Close'], label='SPY Close Price')
plt.title('SPY Close Price Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['VIX_Close'], color='orange', label='VIX Close Price')
plt.title('VIX (Volatility Index) Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data['VIX_Close'], bins=50, kde=True)
plt.title('Distribution of VIX')
plt.show()
```

```python
data['SPY_Returns'] = data['SPY_Close'].pct_change()
plt.figure(figsize=(14,6))
data['SPY_Returns'].plot()
plt.title('SPY Daily Returns Over Time')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['VIX_Close'], y=data['SPY_Returns'])
plt.title('SPY Returns vs VIX')
plt.xlabel('VIX')
plt.ylabel('SPY Daily Returns')
plt.show()
```
