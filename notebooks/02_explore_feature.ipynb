```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
data = pd.read_csv('../data/spy_vix_features.csv', index_col=0, parse_dates=True)
```

```python
print("✅ Feature Data Columns")
print(data.columns)

print("\n✅ Feature Data Head")
print(data.head())
```

```python
plt.figure(figsize=(14,7))
plt.plot(data['SPY_Close'], label='SPY Close')
plt.plot(data['MA50'], label='50-day MA', linestyle='--')
plt.plot(data['MA200'], label='200-day MA', linestyle='--')
plt.title('SPY Close vs Moving Averages')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['RSI'], color='purple')
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.title('RSI Indicator')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['MACD'], color='teal')
plt.title('MACD Histogram (Momentum Signal)')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['Realized_Volatility'], color='orange')
plt.title('10-Day Rolling Realized Volatility')
plt.show()
```

```python
plt.figure(figsize=(14,6))
(data['Trend'] == 'Up').astype(int).plot()
plt.title('Trend Regime (1 = Up, 0 = Down)')
plt.show()
```

```python
plt.figure(figsize=(14,6))
(data['Volatility_Regime'] == 'High').astype(int).plot(color='red')
plt.title('Volatility Regime (1 = High, 0 = Normal)')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data[['SPY_Close', 'RSI', 'MACD', 'Realized_Volatility']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Features')
plt.show()
```
